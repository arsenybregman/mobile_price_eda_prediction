```
import numpy as np
import pandas as pd
```

```
df = pd.read_csv('mobile.csv')
```

```
df.head()
```

```
df.duplicated().sum()
```

```
print(df[df.duplicated()])
```

```
df.isnull().sum()
```

```
del df['img']
```

SIM card

```
df['sim'].unique()
```

```
data = df.copy()
```

```
data['sup_3G'] = [1 if '3G' in x else 0 for x in data['sim']]
data['sup_4G'] = [1 if '4G' in x else 0 for x in data['sim']]
data['sup_5G'] = [1 if '5G' in x else 0 for x in data['sim']]
data['nfc'] = [1 if 'NFC' in x else 0 for x in data['sim']]
```

```
del data['sim']
```

```
data.info()
```

Processor

```
data['processor'].unique()[0:10]
```

```
import re 
```

```
data['proc_name'] = data['processor'].str.extract(r'(\w+)', flags=re.IGNORECASE)
```

```
data['proc_name'].unique()
```

```
data['proc_GHz'] = [
    x.replace(' ', '').split(',')[-1] if isinstance(x, str) and '\u2009' in x else None
    for x in data['processor']
]
data['proc_GHz'] = data['proc_GHz'].str.partition('\u2009')[0].astype('float')
#data['proc_GHz'] = pd.to_numeric(data['proc_GHz'], errors='coerce', downcast='float')
data.loc[data['processor'].str.contains('MHz', na=False), 'proc_GHz'] /= 1000
```

```
data['proc_GHz'].unique()
```

```
del data['processor']
```

Storage

```
data['storage'][0:10]
```

```
data['ram'] = data['storage'].str.extract(r'(\d+\u2009\w+ RAM)', flags=re.IGNORECASE)
data['ram'] = [
    float(x.split('\u2009')[0]) / 1024 if isinstance(x, str) and 'MB' in x else 
    float(x.split('\u2009')[0]) if isinstance(x, str) else None 
    for x in data['ram']
]
```

```
data['ram'].unique()
```

```
data['rom'] = data['storage'].str.extract(r'(\d+\u2009\w+ inbuilt)', flags=re.IGNORECASE)
data['rom'] = [
    float(x.split('\u2009')[0]) * 1024 if isinstance(x, str) and 'TB' in x else 
    float(x.split('\u2009')[0]) if isinstance(x, str) else None 
    for x in data['rom']
]
```

```
data['rom'].unique()
```

Battery

```
data['battery'][0:10]
```

```
data['bat_size'] = data['battery'].str.extract(r'(\d+)\u2009\w+', flags=re.IGNORECASE).astype('float')
#data['bat_size'] = pd.to_numeric(data['bat_size'], errors='coerce', downcast='integer')

data['bat_size'].unique()
```

Display

```
data['display'].unique()
```

Inches

```
data['inches'] = data['display'].str.extract(r'(\d+\.?\d*)').astype('float')
```

display size

```
data['disp_size'] = data['display'].str.extract(r'(\d+\.?\d*\u2009\w+\u2009\d+)')
```

```
data['disp_size'] = [
    int(x.split('\u2009')[0]) * int(x.split('\u2009')[-1]) if isinstance(x, str) else None
    for x in data['disp_size']
]
```

display hz

```
data['disp_hz'] = data['display'].str.extract(r'(\d+\.?\d*) Hz').astype('float')
```

```
data['disp_hz'].unique()
```

```
del data['display']
del data['storage']
del data['battery']

data.info()
```

Camera

```
data['camera'].unique()
```

```
camera = data['camera'].str.partition('&', expand=True)
```

```
camera
```

```
camera[2].unique()
```

```
def rear_search(line):
    line = str(line)
    if 'Dual' in line:
        return 2
    elif 'Triple' in line:
        return 3
    return 1

camera[1] = camera[0].apply(rear_search)
camera[2] = camera[2].str.extract(r'(\d+)\s*MP', flags=re.IGNORECASE).astype(float)
camera[0] = camera[0].str.extract(r'(\d+)\s*MP', flags=re.IGNORECASE).astype(float)
```

```
data['camera'] = camera[0]
data['front_camera'] = camera[2]
data['camera_amount'] = camera[1]
```

```
data.info()
```

External memory

```
data['memoryExternal'].unique()
```

```
ext_memory = data['memoryExternal'].str.split('\u2009', expand=True)
ext_memory[2] = ext_memory[0].str.extract(r'(\d+)', flags=re.IGNORECASE).astype('float')
ext_memory.loc[ext_memory[1] == 'TB', 2] *= 1024
ext_memory[2] = ext_memory[2].round(2)
ext_memory[2].unique()
```

```
# Fixing the TypeError by checking for None instead of float in the list comprehension

data['ext_memory'] = ext_memory[2]
data['is_ext_memory'] = [
    0 if isinstance(x, str) and 'Not' in x 
    else (None if x is None else 1) 
    for x in data['memoryExternal']
]

del data['memoryExternal']

data.is_ext_memory.value_counts()
```

Creator

```
data['creator'] = data['Name'].str.extract(r'(\w+)')
del data['Name']
data['creator'].unique()
```

```
del data['tag']
del data['version']
del data['fm']
del data['proc_name']
```

```
data.round({'proc_GHz': 2, 'inches': 2})
data.head()
```

```
data.isnull().sum()
```

```
data['inches'].unique()
```

```
data['inches'] = data['inches'].astype('float')
```

```
data.dropna(subset=['bat_size', 'inches', 'disp_size', 'camera'], inplace=True)
```

```
data.loc[data['creator'] == 'Oppo', 'creator'] = 'OPPO'
```

```
data.info()
```

```
line = ['disp_size', 'ext_memory', 'creator']
for x in line:
    data[x] = data[x].astype('category')
```

```
data.to_csv('mobile_clean.csv', index=False)
```
